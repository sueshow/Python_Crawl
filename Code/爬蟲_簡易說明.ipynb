#### 方法一：request.read

* [Python 網路爬蟲 Web Crawler 基本教學 By 彭彭](https://www.youtube.com/watch?v=9Z9xKWfNo7k)
* [Python 網路爬蟲 Web Crawler 教學 - Cookie 操作實務 By 彭彭](https://www.youtube.com/watch?v=BEA7F9ExiPY)
* [Python 網路連線程式、公開資料串接 By 彭彭](https://www.youtube.com/watch?v=sUzR3QVBKIo)

說明：
* Cookie：網站存放在瀏覽器的一小段內容
* 與伺服器互動：連線時，Cookie 放在 Request Headers 中送出
* 透過「檢視原始碼」→「Application」→「Cookies」觀察重要的 Cookie

```python
import urllib.request as req
from bs4 import BeautifulSoup 
```

```python
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

# 建立一個 Request 物件，附加 Headers
header = {
           'cookie': 'over18=1',
           'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'
}

request = req.Request(url, headers=header)

# 抓取網頁原始碼
with req.urlopen(request) as response:
    data = response.read().decode('utf-8')

#print(data)
```

```python
# 解析原始碼，取得每篇文章的標題
# 讓 beautifulsoup協助我們解析 HTML 格式文件
root = BeautifulSoup(data, 'html.parser')
```

##### 找尋單一標籤

```python
# 尋找 class='title' 的 div 標籤
titles = root.find('div', class_='title')
print(titles.a.string)
```

##### 找尋所有標籤

```python
# 尋找所有 class='title' 的 div 標籤
titles = root.find_all('div', class_='title')
for title in titles:
    if title.a != None: #如果標題包含 a標籤(沒有被刪除)→印出來
        print(title.a.string)
```

##### 抓取下一頁的連結

```python
# 找到內文是「‹ 上頁」的 a 標籤
nextlink= root.find('a', string='‹ 上頁')
print(nextlink)
print(nextlink['href'])
```

##### 整理程式

```python
def getData(url):
#     url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

    # 建立一個 Request 物件，附加 Headers
    header = {
               'cookie': 'over18=1',
               'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'
    }

    request = req.Request(url, headers=header)

    # 抓取網頁原始碼
    with req.urlopen(request) as response:
        data = response.read().decode('utf-8')

    # 解析原始碼，取得每篇文章的標題
    # 讓 beautifulsoup協助我們解析 HTML 格式文件
    root = BeautifulSoup(data, 'html.parser')

    # 尋找所有 class='title' 的 div 標籤
    titles = root.find_all('div', class_='title')
    for title in titles:
        if title.a != None: #如果標題包含 a標籤(沒有被刪除)→印出來
            print(title.a.string)

    # 找到內文是「‹ 上頁」的 a 標籤
    nextlink= root.find('a', string='‹ 上頁')
    print('='*20)
    return(nextlink['href'])
    
```

```python
# 主程序
# 抓取多頁面的標題
pageURL = 'https://www.ptt.cc/bbs/Gossiping/index.html'
count = 0
while count<3:
    pageURL = 'https://www.ptt.cc'+getData(pageURL)
    count +=1
```

## 範例四：AJAX+JSON

* [Python 網路爬蟲 Web Crawler 教學 - AJAX / XHR 網站技術分析實務 By 彭彭](https://www.youtube.com/watch?v=IMOUf4BYTG8)

說明：
* AJAX：網頁前端的 JavaScript 程式技術，通常第1次請求不帶資料，且透過多次請求取得資料並產生網頁畫面
* 在網頁原始碼搜尋不到，肉眼看到的標題，分析網頁找到能夠抓資料的網址

### 以 Medium 為例(範例格式已改變)+JSON

```python
import urllib.request as req
import json
```

```python
url = 'https://medium.com/_/api/home-feed'

# 建立一個 Request 物件，附加 Headers
header = {
           'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'
}

request = req.Request(url, headers=header)

# 抓取網頁原始碼
with req.urlopen(request) as response:
    data = response.read().decode('utf-8')  # 根據觀察，取得的資料是 JSON 格式
```

```python
print(data)
```

```python
# 解析 JSON 格式的資料，取得每篇文章的標題
# JSON資料格式：大括號是字典，中括號是列表
# 處理髒資料
data = data.replace('])}while(1);</x>','')
data = json.loads(data)  # 把原始的 JSON 資料解析成字典/列表的表示形式
#print(data)

# 取得 JSON 資料中的文章標題
posts = data['payload']['references']['Post']
for key in posts:
    post = posts[key]
    print(post['title'])
```

## 範例五：AJAX+JSON

* [Python 網路爬蟲 Web Crawler 教學 - Request Data 操作實務 By 彭彭](https://www.youtube.com/watch?v=DSjVHqdbkBw)

* 說明
  * 簡單的請求(不帶額外的資料)：發送請求 → 取得 HTML 網頁 → 發送簡單的請求 → 取得回應資料並產生網頁畫面
  * 複雜的請求(帶額外的資料)：發送請求 → 取得 HTML 網頁 → 發送複雜的請求，並提供相關資料 → 取得回應資料並產生網頁畫面
    * 請求的額外資訊：Request Payload
      * json.dumps(Request_Payload).encode('utf-8')
    * content-type：application/json
    * user-agent
* 關鍵：認出網站的 AJAX 模式，並懂得發出複雜的請求
* Request Data：請求附帶的資料，又稱Request Payload、Request Body

### 以 kkday 為例+JSON

* [三個你不知道的Python網頁爬蟲秘密](https://www.learncodewithmike.com/2020/10/scraping-ajax-websites-using-python.html)

```python
import pandas as pd

import requests
import json
```

```python
city_name = '新竹市'
page = 1 #1-4
url = r'https://www.kkday.com/zh-tw/product/ajax_productlist/?page={page}&keyword={city_name}&sort=prec'
# 取得傳入城市的所有一日遊票券
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=header)
#print(response)

# 資料
activities = response.json()['recommend_productlist']
#print(activities)

#tot_page = 

result =[]
for activity in activities:
    # 票券名稱
    title = activity['name']
    # 票券詳細內容連結
    link = activity['url']
    # 票券價格
    price = activity['price']
    # 最早可使用日期
    booking_date = activity['earliest_sale_date']
    # 評價
    star = activity['rating_star']
    # 可免費退費
    free_refund = activity['free_refund_policy']
    result.append(dict(title=title, link=link, price=price, booking_date=booking_date, star=star, free_refund=free_refund, source='KKday'))
#print(result)
#print(result.key)
df = pd.DataFrame(result)
df
```

### 物件化

```python
# KKday網站
class Kkday:
 
    #建構式
    def __init__(self, city_name):
        self.city_name = city_name  # 城市名稱屬性
        
    def scrape(self):
        result = []  # 回傳結果
        
        if self.city_name:  # 如果城市名稱非空值
            url = r'https://www.kkday.com/zh-tw/product/ajax_productlist/?keyword={self.city_name}&sort=prec'
            header = {
                      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'
                     }
            
            # 取得傳入城市的所有一日遊票券
            response = requests.get(url, headers=header)
            
            # 資料
            activities = response.json()['recommend_productlist']
            
            for activity in activities:
                # 票券名稱
                title = activity['name']
                # 票券詳細內容連結
                link = activity['url']
                # 票券價格
                price = activity['price']
                # 最早可使用日期
                booking_date = activity['earliest_sale_date']
                # 評價
                star = activity['rating_star']
                # 可免費退費
                free_refund = activity['free_refund_policy']
                
                result.append(
                    dict(title=title, link=link, price=price, booking_date=booking_date, star=star, free_refund=free_refund, source='KKday'))
 
        return result
```

```python
demo = Kkday('花蓮')
#print(demo.scrape())
df = pd.DataFrame(demo.scrape())
df
```
