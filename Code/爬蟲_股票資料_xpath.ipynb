## yahoo!股市 每日技術分析

* 開啟「Network」後先按"清除圖示"(Clear)=>重整網頁=>按下"搜尋圖示"(Search)
* 搜尋特定數值=>Headers > General > Request URL 就是我們需要的網址
* 範例網址：https://tw.quote.finance.yahoo.net/quote/q?type=ta&perd=d&mkt=10&sym=2330&v=1&callback=jQuery1113009105977637450136_1582441974703&_=1582441974704
* 參數從q?後開始(在 element 按右鍵=>展開資料)
  * type：mem、ta(每天成交價資訊)
  * perd：5分線(5m)、10分線(10m)、30分線(30m)、60分線(60m)、日線(d)、週線(w)、月線(m)
  * mkt：
  * sym：股票代碼(範例為2330)
  * v：

```python
import requests
url = 'https://tw.quote.finance.yahoo.net/quote/q?type=ta&perd=d&mkt=10&sym=2330&v=1'
response = requests.get(url)
html = response.text
print(html)
```

```python
import json
data = html.split('null(')[-1].split(');')[0]
json_data = json.loads(data)
print(json_data['mkt'], json_data['id'])
```

## [台灣銀行](https://fund.bot.com.tw/z/index.htm)

找出 url：用開啟「Network」的方法找
* 範例網址：https://fund.bot.com.tw/Z/ZC/ZCW/CZKC1.djbcd?a=2330&b=D&c=1440
* 參數：
  * a：股票代碼
  * b：天(D)
  * c：天數(最大天數為1440)

```python
import requests 
url = 'https://fund.bot.com.tw/Z/ZC/ZCW/CZKC1.djbcd?a=2330&b=D&c=1440' ### 找出 url
resp = requests.get(url)
html = resp.text
group = html.split(' ')
g_date, g_open, g_high, g_low, g_close, g_volume = group[0:5+1]
g_date = g_date.split(',')
g_open = g_open.split(',')
g_high = g_high.split(',')
g_low = g_low.split(',')
g_close = g_close.split(',')
g_volume = g_volume.split(',')
datas = []
for i in range(0, len(g_date)):
    datas.append([g_date[i], g_open[i], g_high[i], g_low[i], g_close[i], g_volume[i]])
    print(datas[i])
```

## [台灣證交所](https://www.twse.com.tw/zh/page/trading/exchange/STOCK_DAY.html)

* 會阻擋爬蟲
* 用來補齊過往資料
* 從網頁上可找到下載的路徑，即是url

* 範例網址：HTML https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&date=20200223&stockNo=2330
* 參數：
  * response：html、csv
  * date：日期(範例20200223)
  * stockNo：股票代碼(2330)

### HTML

```python
import requests 
from lxml import etree
url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&date=20220223&stockNo=2330' ### 找出 url
resp_html = requests.get(url)
html = resp_html.text
page = etree.HTML(html)
datas = []
for tr in page.xpath('//tr'):
    row = []
    for td_text in tr.xpath('.//td//text()'):
        row.append(td_text.strip())
    if row:
        datas.append(row)
        print(row)
```

### CSV

```python
import requests
url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=csv&date=20200223&stockNo=2330' ### 找出 url
resp = requests.get(url)
html = resp.text
html = html.replace('\r','')
for row in html.split('\n'):
    datas = row.split('","')
    if len(datas)==9:
        datas[0] = datas[0].replace('"','')
        datas[-1] = datas[-1].replace('"','')
        print(datas)
```

### 函式
* 以 台灣銀行的爬蟲為例，現在有兩隻股票 2330, 2317 要解析

```python
import requests
def get_html(stock_id, sdate='20200201', source=''):
    if source=='fund_bot':
        url = 'https://fund.bot.com.tw/Z/ZC/ZCW/CZKC1.djbcd?a={}&b=D&c=1440'.format(stock_id)
    elif source=='twse':
        url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&date={}&stockNo={}'.format(sdate,stock_id)
    else:
        return ''                                 ##函式的斷點
    resp = requests.get(url)
    html = resp.text
    return html

def parse_fundbot(html):                        ##台灣銀行
    group = html.split(' ')
    g_date, g_open, g_high, g_low, g_close, g_volume = group[0:5+1]
    g_date = g_date.split(',')
    g_open = g_open.split(',')
    g_high = g_high.split(',')
    g_low = g_low.split(',')
    g_close = g_close.split(',')
    g_volume = g_volume.split(',')
    datas = []
    for i in range(0,len(g_date)):
        datas.append([g_date[i], g_open[i], g_high[i], g_low[i], g_close[i], g_volume[i]])
    return datas

def parse_twse_html(html):                       ##台灣證交所
    page = etree.HTML(html)
    datas = []
    for index,tr in enumerate(page.xpath('//tr')):
        if index>1:
            row = []
            for td_text in tr.xpath('.//td//text()'):
                text= td_text.strip().replace(',','')
                if text.count('/')==2:
                    y,m,d = text.split('/')
                    y = int(y)+1911
                    text = '{}/{}/{}'.format(y,m,d)
                row.append(text)
            if row:
                datas.append([row[0], row[3], row[4], row[5], row[6], str(round(float(row[1])/1000))])  #以張數計算，四捨五入
    return datas

for stock_id in [2330,2317]:
    html = get_html(stock_id, source='fund_bot')
    datas = parse_fundbot(html)
    # html = get_html(2330,source='twse')
    # datas = parse_twse_html(html)
    for i in range(0, len(datas)):
        print(stock_id, datas[i])
```
